Q) Make a matrix with only border elements as 1 and rest all as 0

```python
import numpy as np

def border_arrays(a,b):
    if a < 3 or b < 3:
        raise ValueError('dimensions should be greater than 2 on each axis')
    arr = np.ones((a,b))
    arr[1:-1, 1:-1] = 0
    return arr

# a,b = map(int,input('Enter the shape of matrix').split())
# border_arrays(a,b)
```

Q) write softmax function

```python
def softmax(arr):
    if type(arr) is not np.ndarray:
        raise TypeError('requires numpy array')
    if arr.ndim > 1:
        raise TypeError('requires 1D array')
    a = np.exp(arr)
    return a/np.sum(a)

arr = np.random.randint(1,15, 10)
softmax(arr)
```

Q) make a function that can take any number of ndarrays (raise error if not ndarray) and vstack them

```python
def vertical_stack(*v):
    for i in v:
        if type(i) is not np.ndarray:
            raise TypeError('requires numpy array')

    return np.vstack(v)

a = np.arange(10).reshape((2,-1))
b = np.repeat(1,10).reshape((2,-1))

vertical_stack(a,b)
```

Q) replace the elements in the array([1,2,np.nan],[4,2,6],[np.nan,np.nan,5]) with the most common element in the array

```python
from collections import Counter

a = np.array([[1,2,np.nan,6],[4,2,6,6],[np.nan,np.nan,5,3]])

dict = {}
for i in a.ravel():
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1

x = sorted(dict.items(),key=lambda x:x[1])[-1][0] #OR Counter(np.ravel(a)).most_common(1)
a[np.isnan(a)] = x
a, np.nan_to_num(np.ravel(a))
```

Q) Check if the give matrices are broadcastable

```python
def check_broadcast(a,b):
    a = a[::-1]
    b = b[::-1]

    for i,j in zip(a,b):
        if i!=1 and j!=1:
            if i != j:
                return False
    return True

a = np.arange(2*3*3*4).reshape(2,3,3,4)
b = np.arange(2*4*3).reshape(2,4,3)
c = np.arange(2*3*4).reshape(2,3,4)
d = np.arange(3*1*4).reshape(3,1,4)

check_broadcast(a.shape,b.shape),check_broadcast(a.shape,c.shape),check_broadcast(a.shape,d.shape)
# a + c, a + b
# a+d
```

Q) given a matrix (i,j) where i: students, j: subjects and (i,j): marks  
- add a new subject with random marks given to each students (between 50-100)      
- create 2 new students with random marks for each subjects (between 50-100)     
- create a new column representing the total marks of each student for all subjects and sort the matrix by highest total marks

```python
marks_data = np.random.randint(1,100,(4,5))
subject = np.random.randint(50,100, (marks_data.shape[0],1))

marks_data = np.concatenate([marks_data,subject], axis=1) #new subject added

s1 = np.random.randint(50,100,(1,marks_data.shape[1]))
s2 = np.random.randint(50,100,(1,marks_data.shape[1]))

marks_data =  np.concatenate([marks_data, s1,s2]) #2 new subjects added

total = np.sum(marks_data, axis=1, keepdims = True) #total for each student added
marks_data = np.concatenate([marks_data, total], axis=1)

marks_data[marks_data[:,-1].argsort()], np.array(sorted(marks_data, key=lambda x: x[-1]))
```
