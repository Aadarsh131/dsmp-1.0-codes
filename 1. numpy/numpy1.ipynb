## Indexing and slicing

```python
#chain indexing (not recommended)
a, a[0], a[2][1], a[-1][-1]
```

```python
#multidimentional indexing(recommended, much efficient and faster than chain indexing)
a[:,3], a[1:,1:3]
```

```python
print(a[::2, ::3]) #fetch only corner elemnts
a[::2, 1::2], a[::2, ::-1], a[::-1, ::-1], a[::-1][::2,::3]
```

```python
c = np.arange(27).reshape(3,3,3) 
c, c[::2][::, ::, 1], c[::2][::, 1::, 1::], c[:,1:2, 1:2], c[::2, 0, ::2], c[:,:,1]==c[::,::,1] #[blocks, rows, cols] (each element i.e, blocks, rows and cols follows start:end:stepsize)
```

```python
print(a)
for i in a:
    print(i) #prints rowwise

for i in a:
    for j in i:
        print(j, end=' ') #prints elementwise
print("\n")

for i in np.nditer(a):
    print(i, end=' ') #prints elementwise
```
